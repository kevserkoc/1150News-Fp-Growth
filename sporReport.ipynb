```python
import pandas as pd
```

```python
data=pd.read_csv("1150haber.csv")
data
```

```python
spor=data[data["oz39699"]=="spor"]
spor
```

```python
spor = spor.drop(spor.columns[:34], axis=1)
```

```python
spor = spor.drop(spor.columns[8206:39698], axis=1)
```

```python
spor = spor.loc[:, (spor != 0).any(axis=0)]

spor.shape
```

```python
spor
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

r=spor.sum(axis=0).sort_values(ascending=False)[:30]
plt.figure(figsize=(20,10))
s=sns.barplot(x=r.index,y=r.values)
s.set_xticklabels(s.get_xticklabels(), rotation=90)

plt.show()
```
