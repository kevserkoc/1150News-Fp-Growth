```python
import pandas as pd
```

```python
data=pd.read_csv("1150haber.csv")
data
```

```python
ekonomi=data[data["oz39699"]=="ekonomi"]
ekonomi
```

```python
ekonomi.head()
```

```python
ekonomi.shape
```

```python
ekonomi = ekonomi.drop(ekonomi.columns[:34], axis=1)
```

```python
ekonomi = ekonomi.drop(ekonomi.columns[8206:39698], axis=1)
```

```python
ekonomi = ekonomi.loc[:, (ekonomi != 0).any(axis=0)]

ekonomi.shape
```

```python
ekonomi
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

r=ekonomi.sum(axis=0).sort_values(ascending=False)[:30]
plt.figure(figsize=(20,10))
s=sns.barplot(x=r.index,y=r.values)
s.set_xticklabels(s.get_xticklabels(), rotation=90)

plt.show()
```
